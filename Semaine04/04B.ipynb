# Exercice

- Une tonne de poires rapporte 3 keuros et une tonne de pommes 2keuros.
- Une tonne de poires nécessite 2 hectares de surface, 1 m^3 d'eau et 1 kg d'engrais.
- Une tonne de pommes nécessite 1 hectare de surface, 1 m^3 d'eau et 2 kg d'engrais
- On a à notre disposition 8 hectares, 5 m^3 d'eau et 8 kg d'engrais.
- Comment maximiser son revenu?

# Modélisation

- Soient $x$ le nombre de tonnes de pommes à produire et $y$ le nombre de tonnes de poires produites.
- En tant que quantités on a 
$$x\geq 0,\qquad y\geq 0.$$
- La surface de cultivation doit être au plus $8$ hectares donc
$$x+2y \leq 8.$$
- On utilise au plus 5 m^3 d'eau donc
$$x + y \leq 5.$$
- Pour l'engrais on obtient
$$2x + y \leq 8.$$
- Le revenu $R$ à maximiser est donné par
$$R=2x+3y.$$

Au final on cherche
$$\underset{(x,y)\in D}{\max}2x+3y.$$
Avec
$$D = \{(x,y)\in R_+^2\ :\ x+2y \leq 8,\quad x+y\leq 5, \quad 2x+y\leq 8\}$$

# Résolution graphique

- Dessiner $D$.
- Dessiner les isoclines de $R$.
- En déduire la solution.

```python
import sympy as sp
```

```python
sp.init_printing()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
x, y = sp.symbols("x, y")
```

## Domaine

```python
p = sp.plot_implicit(sp.Eq(x, 0), (x, -1, 5), (y, -1, 5), show=False, line_color="red")
p.extend(sp.plot_implicit(sp.Eq(y, 0), (x, -1, 5), show=False, line_color="red"))
p.extend(sp.plot_implicit(sp.Eq(x + 2 * y, 8), show=False, line_color="red"))
p.extend(sp.plot_implicit(sp.Eq(x + y, 5), show=False, line_color="red"))
p.extend(sp.plot_implicit(sp.Eq(2* x + y, 8), show=False, line_color="red"))
p.show()
```

## Isoclines

```python
R = 2 * x + 3 * y
R
```

```python
p1 = sp.plot_implicit(sp.Eq(R, 0), (x, -1, 3), show=False)
for niveau in range(1, 5):
    p1.extend(sp.plot_implicit(sp.Eq(R, niveau), show=False))

p1.show()
```

## Résolution graphique

```python
def solution(niveaux):
    p = sp.plot_implicit(sp.Eq(R, 0), (x, -1, 5), (y, -1, 5), show=False)
    for niveau in niveaux:
        p.extend(sp.plot_implicit(sp.Eq(R, niveau), show=False))
    
    
    p.extend(sp.plot_implicit(sp.Eq(x, 0), show=False, line_color="red"))
    p.extend(sp.plot_implicit(sp.Eq(y, 0), show=False, line_color="red"))
    p.extend(sp.plot_implicit(sp.Eq(x + 2 * y, 8), show=False, line_color="red"))
    p.extend(sp.plot_implicit(sp.Eq(x + y, 5), show=False, line_color="red"))
    p.extend(sp.plot_implicit(sp.Eq(2* x + y, 8), show=False, line_color="red"))
    p.show()
```

```python
solution((1, 2, 3))
```

```python
solution([i for i in range(1, 14)])
```

On voit que l'isocline de niveau maximal touchant $D$ correspond à la valeur $13$ et touche $D$ au point 
$x=2,\ y=3$

# Remarque

- La solution graphique est limité au cas de $2$ ou $3$ variables, il faudra chercher autre chose pour une solution générale.
- Le domaine de contrainte était un polygone, on verra que c'est le cas en toute généralité.
- On note que le maximum a été trouvé sur un sommet du polygone de contrainte, là encore ce sera vrai en toute généralité.

Mettez le problème précédent sous forme standard :

## Solution

En posant
$$
\begin{cases}
u = 8 - (x + 2y)\\
v = 5 - (x + y)\\
w = 8 - (2x + y)
\end{cases}
$$
On se ramène au problème

$$
\underset{(x,y,u,v,w)\in C}{\max}R(x,y,u,v,w)
$$
avec
$$R(x,y,u,v,w)= 2x+3y,$$
$$C = \{(x,y,u,v,w)\in \mathbb{R}_+^5\ :\ x+2y+u=8,\ x+y+v=5,\ 2x+y+w=8\}.$$

en introduisant la matrice
$$
M =
\begin{pmatrix}
1 & 2 & 1 & 0 & 0 \\
1 & 1 &  0 & 1 & 0\\
2 & 1 & 0 & 0 & 1
\end{pmatrix}
$$
et le vecteur
$$d = (8, 5,8)$$
la contrainte devient
$$C=\{z \in \mathbb{R}_+^5\ :\ Mz=d\}$$
